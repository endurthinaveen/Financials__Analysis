# 2.	Remove '$' sign and '-' from all columns where they are present

```python
def clean_value(value):
    if isinstance(value, str):
        value = value.replace('$', '').replace('-', '').replace(",", "").replace(" ","")
    return value
df = df.applymap(clean_value)
```

1. **Sales Performance Analysis**: Analyze sales trends, best-selling products, and highest revenue-generating segments.

```python

columns_to_convert = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']
for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column].str.replace('[\$,]', '', regex=True), errors='coerce')
print(df.head())
```

```python

df['Date'] = pd.to_datetime(df['Date'])
df_monthly = df.resample('M', on='Date').sum()
plt.figure(figsize=(10, 6))
plt.plot(df_monthly.index, df_monthly['Sales'], label='Monthly Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Monthly Sales Trends')
plt.legend()
plt.grid()
plt.show()

product_sales = df.groupby('Product')['Sales'].sum()
best_selling_products = product_sales.nlargest(5)
print("Best-Selling Products:")
print(best_selling_products)


segment_revenue = df.groupby('Segment')['Sales'].sum()
revenue_generating_segments = segment_revenue.nlargest(3)
print("Revenue-Generating Segments:")
print(revenue_generating_segments)
```

# 6. **Segmentation Strategy**: Identify the most lucrative segments and tailor marketing strategies accordingly

```python

numeric_columns = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

segment_profit = df.groupby('Segment')['Profit'].sum()
most_lucrative_segments = segment_profit.nlargest(3)
print("Most Lucrative Segments:")
print(most_lucrative_segments)

for segment in most_lucrative_segments.index:
    segment_data = df[df['Segment'] == segment]
    products_sold = segment_data['Product'].unique()
    countries = segment_data['Country'].unique()
    discount_bands = segment_data['Discount Band'].unique()
    print(f"Segment: {segment}")
    print(f"Products Sold: {products_sold}")
    print(f"Countries: {countries}")
    print(f"Discount Bands: {discount_bands}")
    print("=" * 30)

plt.figure(figsize=(10, 6))
segment_profit.plot(kind='bar', color='purple')
plt.xlabel('Segment')
plt.ylabel('Total Profit')
plt.title('Segment Profitability Analysis')
plt.xticks(rotation=0)
plt.grid()
plt.show()
```

well-structured and informative, allowing you to better understand the characteristics of the most profitable segments

# 8. **Month and Year Trends**: Study how sales, revenue, and profit change across months and years.

# 9. **Cost Management**: Explore ways to optimize costs, improve profitability, and reduce discounts.

```python
numeric_columns = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

df['Profit Margin'] = df['Profit'] / df['Sales']

low_margin_products = df[df['Profit Margin'] < 0.1]
segment_discounts = df.groupby('Segment')['Discounts'].mean()
high_discount_segments = segment_discounts[segment_discounts > 0.1]
discount_effectiveness = df.groupby('Discounts')[['Sales', 'Profit']].mean()
product_profitability = df.groupby('Product')['Profit'].sum()
segment_profitability = df.groupby('Segment')['Profit'].sum()
discount_effectiveness = df.groupby('Discounts')[['Sales', 'Profit']].mean()
print("Products with Low Profit Margins:")
print(low_margin_products[['Product', 'Profit Margin']])
print("\nSegments with High Discounts:")
print(high_discount_segments)
print("\nDiscount vs. Profitability Analysis:")
print(discount_effectiveness)
print("\nProduct Profitability Analysis:")
print(product_profitability.nlargest(5))
print("\nSegment Profitability Analysis:")
print(segment_profitability.nlargest(3))
```
